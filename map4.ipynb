```python
def datetime(date,format_string):
    return pd.to_datetime(date,format= format_string,errors='coerce')

eventsData["timestamp"]= datetime(eventsData["timestamp"],"%Y-%m-%dT%H:%M:%S")

```

```python
eventsData["month"] = eventsData["timestamp"].map(lambda x: x.month)
eventsData["day"] = eventsData["timestamp"].map(lambda x: x.day)
```

```python
eventsData.head()
```

```python
last_15 = eventsData
```

```python
last_15.head()
```

```python
last_15 = last_15[last_15["event"] == "checkout"]
```

```python
last_15 = last_15[last_15["month"] == 5]
```

```python
last_15 = last_15[last_15["day"] >14]
```

```python
last_15_product = last_15[["person","model"]]
```

```python
last_15_product.head()
```

```python
last_15_product = last_15_product.dropna() 
```

```python
last_15_product["marca"] = last_15_product["model"].map(lambda x:"apple" if(x.split(" ")[0] in ["iphone","ipad"]) else x.split(" ")[0])
```

```python
last_15_product["version"] = last_15_product["model"].map(lambda x:" ".join(x.split(" ")[0:2]) if(x.split(" ")[0] in ["iphone","ipad"]) else " ".join(x.split(" ")[1:3]))
```

```python
last_15_product.head()
```

```python
last_15_product = last_15_product.drop("model",axis=1)
```

```python
"samsung" in prob_marcas.keys()
```

```python
last_15_product["prob_marca"] = last_15_product["marca"].map(lambda x:prob_marcas[x] if(x in prob_marcas.keys()) else 0)
```

```python
last_15_product["prob_model"] = last_15_product["version"].map(lambda x:prob_model[x] if(x in prob_model.keys()) else 0)
```

```python
last_15_product.head()
```

```python
last_15_product = last_15_product[["person","prob_marca","prob_model"]]
```

```python
last_15_product.head()
```

```python
last_15_product["cant_conversiones"] = 1
```

```python
last_15_product = last_15_product.groupby("person",as_index = False).sum()
```

```python
last_15_product.head()
```

```python
last_15_product["prob_marca_en_base_checkout"] = last_15_product["prob_marca"] / (last_15_product["cant_conversiones"]+1)
```

```python
last_15_product.head()
```

```python
last_15_product["prob_model_en_base_checkout"] = last_15_product["prob_model"] /(last_15_product["cant_conversiones"]+1)
```

```python
last_15_product.head()
```

```python
last_15_product = last_15_product[["person","prob_marca_en_base_checkout","prob_model_en_base_checkout"]]
```

```python
last_15_product.head()
```
