```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.naive_bayes import GaussianNB
import sklearn
```

```python
total_train = pd.read_csv('../setTraining.csv')
```

```python
print('The scikit-learn version is {}.'.format(sklearn.__version__))
```

```python

from imblearn.datasets import fetch_datasets
from kmeans_smote import KMeansSMOTE

#datasets = fetch_datasets(filter_data=['oil'])
X, y = total_train.drop(["label","person"]), total_train["label"]

[print('Class {} has {} instances'.format(label, count))
 for label, count in zip(*np.unique(y, return_counts=True))]

kmeans_smote = KMeansSMOTE(
    kmeans_args={
        'n_clusters': 100
    },
    smote_args={
        'k_neighbors': 10
    }
)
X_resampled, y_resampled = kmeans_smote.fit_sample(X, y)

[print('Class {} has {} instances after oversampling'.format(label, count))
 for label, count in zip(*np.unique(y_resampled, return_counts=True))]
```

```python
total_train["label"].value_counts()
```

```python
total_train = total_train.sort_values(by = "label",ascending = False).head(6000)
```

```python
total_train["label"].head()
```

```python
total_train = total_train.drop( ['person','Unnamed: 0'],axis=1)
```

```python
total_train.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
for column in total_train.columns:
    if total_train[column].dtype == type(object):
        total_train[column] = le.fit_transform(total_train[column].astype(str))

total_train.sample(5)
```

```python
total_train.count()
```

```python
total_train.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(total_train.drop(['label'], axis=1).values, total_train['label'], test_size=0.30, random_state=75)
```

```python
clf = GaussianNB()
```

```python
clf.fit(X_train, y_train) 
```

```python
#Para guardar el modelo en un archivo
from sklearn.externals import joblib
joblib.dump(clf, 'GaussianNBFinal.pkl') 
```

```python
#Para cargar el modelo
from sklearn.externals import joblib
clf = joblib.load('GaussianNBFinal.pkl') 
```

```python
clf.predict(X_test)
```

```python
#Pruebo el score para ver si realiza overfitting
clf.score(X_test,y_test)
```
