```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier

%matplotlib inline
```

```python
userTraining = pd.read_csv('../setUsersTrainingNormalizado.csv')
```

```python
userTraining.head()
```

```python
userTraining= userTraining.drop(['Unnamed: 0'],axis=1)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
for column in userTraining.columns:
    if userTraining[column].dtype == type(object):
        userTraining[column] = le.fit_transform(userTraining[column].astype(str))

userTraining.sample(5)
```

```python
columns = userTraining.columns
```

```python
userTraining = userTraining.drop("person",axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(userTraining.drop(['label'], axis=1).values, userTraining['label'], test_size=0.30, random_state=75)
```

```python
clf = GradientBoostingClassifier(n_estimators=100, learning_rate=0.5,random_state=123).fit(X_train, y_train)
```

```python
clf.predict(X_test)
```

```python
clf.score(X_test,y_test)
```
